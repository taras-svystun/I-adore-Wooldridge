```python
import wooldridge
import statsmodels.api as sm
import statsmodels.formula.api as smf
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import numpy as np
import scipy.stats as st
```

## Task 1

```python
data = wooldridge.data('BWGHT')
data
```

```python
correlation = data.corr()
print(correlation['cigs']['faminc'])

fig = px.imshow(correlation, text_auto=True,
               title='Correlation among all variables')
fig.show()
```

```python
restricted_model = smf.ols(formula='bwght ~ cigs', data=data)
restricted_fitted = restricted_model.fit()
beta0 = round(restricted_fitted.params[0], 2)
beta1 = round(restricted_fitted.params[1], 2)
print(f'Restricted model: bwght = {beta0}\
 {beta1} * cigs + u')
print(f'R-squared of restricted = {restricted_fitted.rsquared}')
print(f'Number of observations of restricted = {restricted_model.nobs}')
print(f'adjusted R-squared of restricted = {restricted_fitted.rsquared_adj}')
```

```python
unrestricted_model = smf.ols(data=data, formula='bwght ~ cigs + faminc')
unrestricted_fitted = unrestricted_model.fit()
beta0 = round(unrestricted_fitted.params[0], 2)
beta1 = round(unrestricted_fitted.params[1], 2)
beta2 = round(unrestricted_fitted.params[2], 2)
print(f'Unrestricted model: bwght = {beta0}\
 {beta1} * cigs + {beta2} * faminc + u')
print(f'R-squared of unrestricted = {unrestricted_fitted.rsquared}')
print(f'Number of observations of unrestricted = {unrestricted_model.nobs}')
print(f'adjusted R-squared of unrestricted = {unrestricted_fitted.rsquared_adj}')
```

## Task 2

```python
data = wooldridge.data('HPRICE1')
data
```

```python
figure = make_subplots(rows=2, cols=1)
figure.add_trace(
    go.Scatter(x=data['sqrft'], y=data['price'], mode='markers', name='price on sqrft'),
    row=1, col=1
)
figure.add_trace(
    go.Scatter(x=data['bdrms'], y=data['price'], mode='markers', name='price on bdrms'),
    row=2, col=1
)
figure.show()
```

```python
data = sm.add_constant(data)
y, X = data['price'], data[['const', 'sqrft', 'bdrms']]
model = sm.OLS(y, X)
fitted = model.fit()
print(fitted.summary())
```

```python
print(f'Price increase with 1 additional bedroom, with square footage being constant \
is ${round(fitted.params[2] * 1, 2)}K')
print(f'Price increase with 1 additional bedroom, with 140 square feet \
is ${round(140 * fitted.params[1] + fitted.params[2] * 1, 2)}K')
print(fitted.rsquared ** 2, '% of Variation in price is explained by sqrft and bdrms')
test = np.column_stack([[1], [2438], [4]])
print(f'Predictions for the first house in the dataset (sqrft=2,438; bdrms=4) is \
{round(fitted.predict(test)[0], 2)}')
```

```python
data = wooldridge.data('DISCRIM')
data
```

```python
m_prpblck, m_income = data[['prpblck', 'income']].mean()
std_prpblck, std_income = data[['prpblck', 'income']].std()
m_prpblck, m_income, std_prpblck, std_income
```

```python
model = smf.ols(data=data, formula='psoda ~ prpblck + income')
fitted = model.fit()
print(fitted.summary())
```

```python
restricted_model = smf.ols(data=data, formula='psoda ~ prpblck')
restricted_fitted = restricted_model.fit()
print(restricted_fitted.summary())
```

```python
model = smf.ols(data=data, formula='np.log(psoda) ~ prpblck + np.log(income)')
fitted = model.fit()
print(f'Estimated increase in psoda if prpblck c.p. increases by 0.2 is \
{round(100 * fitted.params[1] * 0.2, 2)}%')
print(fitted.summary())
```

```python
model = smf.ols(data=data, formula='np.log(psoda) ~ prpblck + np.log(income) + prppov')
fitted = model.fit()
corr = np.corrcoef(np.log(data['income']), 100 * data['prppov'])
print(f'Correlation between log(income) and prppov is {corr}')
print(fitted.summary())
```

```python
df = data.copy()
df['logical_educ'] = df['educ'].apply(lambda educ: 'more than 12' if educ > 12 else 'less than 12')
fig = px.histogram(data_frame=df, x='educ', color='logical_educ')
fig.show()
```

```python
figure = go.Figure()
figure.add_trace(go.Scatter(x=data['educ'], y=data['fatheduc'], mode='markers', name='father'))
figure.add_trace(go.Scatter(x=data['educ'], y=data['motheduc'], mode='markers', name='mother'))
fig = px.scatter(data_frame=data, x='educ', y='fatheduc', trendline='ols', trendline_color_override='blue')
trendline = fig.data[1]
figure.add_trace(trendline)
fig = px.scatter(data_frame=data, x='educ', y='motheduc', trendline='ols', trendline_color_override='red')
trendline = fig.data[1]
figure.add_trace(trendline)
figure.update_layout(title="Educ depending on mother and father educ",
    xaxis_title="parent educ",
    yaxis_title="my educ")
figure.show()
```

```python
model = smf.ols(data=data, formula='educ ~ motheduc + fatheduc')
fitted = model.fit()
print(fitted.summary())
```

```python
model = smf.ols(data=data, formula='educ ~ motheduc + fatheduc + abil')
fitted = model.fit()
print(fitted.summary())
```

```python
from scipy.optimize import fmin
model = smf.ols(data=data, formula='educ ~ motheduc + fatheduc + abil + np.square(abil)')
fitted = model.fit()
beta0, beta1, beta2, beta3, beta4 = fitted.params

def regression(abil):
    return beta3*abil + beta4*abil**2

print(fmin(regression, x0=0))
print(round(len(data[data['abil'] < -3.967]) * 100 / len(data), 2), '%')
print(fitted.summary())
```

```python
x_min, x_max = (data['abil'].min(), data['abil'].max())
x = np.linspace(x_min, x_max, 10 ** 4)
def regression(abil):
    return beta0 + beta1 * data.motheduc.mean() + beta2 * data.fatheduc.mean() + \
    beta3 * abil + beta4 * abil ** 2
y = list(map(regression, x))
fig = px.scatter(x=x, y=y)
fig.show()
```
