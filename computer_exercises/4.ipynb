## Task1

```python
data = wooldridge.data('VOTE1')
data
```

```python
model = smf.ols(data=data, formula='voteA ~ np.log(expendA) + np.log(expendB) + prtystrA')
fitted = model.fit()
print(fitted.summary())
```

#### Now, let's test the hypothesis:
    
$ H_0 : \beta_1 = - \beta_2.$
   
To do this, let's set $ \theta = \beta_1 + \beta_2 $, or $ \beta_2 = \theta - \beta_1 $.

Now our regression will look like:

$voteA = \beta_0 + \beta_1 \cdot [ log(expendA) - log(expendB) ] + \theta_1 \cdot log(expendB) + \beta_3 \cdot prtystrA + u$

```python
df = data[ 'voteA$expendA$expendB$prtystrA'.split('$') ].copy()
df[['expendA', 'expendB']] = df[['expendA', 'expendB']].apply(np.log)
df['log_diff'] = df['expendA'] - df['expendB']

model = smf.ols(data=df, formula='voteA ~ log_diff + expendB + prtystrA')
fitted = model.fit()
print(fitted.summary())
```

As we can see, $\theta$ is not significant, so we cannot reject the null hypothesis.

## Task 11

```python
data = wooldridge.data('HTV')
data.head()
```

```python
model = smf.ols(data=data, formula='educ ~ motheduc + fatheduc + abil + np.square(abil)')
fitted = model.fit()
fitted.summary()
```

Now I'll test hypothesis: $H_0:$ relationship between educ and abil is linear vs $H_1:$ relationship is quadratic.

```python
fitted.f_test(np.array(((0, 0, 0, 1, 0), (0, 0, 0, 0, 1)))).summary()
```

We see, that relationship is quadratic.

```python
fitted.f_test('(abil=0), (np.square(abil)=0)').summary()
```
